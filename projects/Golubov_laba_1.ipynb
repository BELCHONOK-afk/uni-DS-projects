Задача: написать нейросеть-классификатор, определяющую тип звезды. <br>

$X =\{\langle Temperature \rangle, \langle Luminosity \rangle, \langle Radius \rangle, \langle Absolute\ magnitude \rangle \} $ <br>
$Y  = \{ Star\ type: [0,1,2,3,4,5]\}$

```python
import tensorflow as tf
from tensorflow import keras 
import numpy as np
import pandas as pd
from sklearn import preprocessing, model_selection
```

```python
data = pd.read_csv("/Users/karantovanatolij/Documents/Unis' rubbish/Golubov/classes.csv")
data.tail()
data.dropna()
```

```python
x = np.array(data.drop(['Star type', 'Star color', 'Spectral Class'],1))
y = np.array(data['Star type'], dtype = 'float')
#y.shape = (len(y), 1)
#print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y,train_size = 0.7)
```

```python
x_f_train = preprocessing.scale(x_train)
x_f_test = preprocessing.scale(x_test)
y_f_train = y_train
y_f_test = y_test
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())# the input layer
model.add(tf.keras.layers.Dense(10, activation=tf.nn.relu))# hiden layer 
model.add(tf.keras.layers.Dense(30, activation=tf.nn.relu))

model.add(tf.keras.layers.Dense(6, activation=tf.nn.softmax))


model.compile(optimizer= "SGD",
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_f_train, y_f_train, epochs=100)
```

```python
initializer = tf.keras.initializers.RandomNormal(mean = 0.0, stddev = 0.05)
initb = tf.keras.initializers.Constant(value=1e-3)

model2 = tf.keras.Sequential()
model2.add(tf.keras.layers.Flatten())
model2.add(tf.keras.layers.Dense(10, activation= tf.nn.relu, kernel_initializer=initializer, bias_initializer=initb))
model2.add(tf.keras.layers.Dense(30, activation= tf.nn.relu, kernel_initializer=initializer, bias_initializer=initb))
model2.add(tf.keras.layers.Dense(6, activation=tf.nn.softmax))

model2.compile(optimizer='SGD',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

model2.fit(x_f_train, y_f_train, epochs=100)
```

```python
initializer = tf.keras.initializers.GlorotNormal()
initb = tf.keras.initializers.Constant(value = 1e-3)

model3 =tf.keras.Sequential()
model3.add(tf.keras.layers.Flatten())
model3.add(tf.keras.layers.Dense(100, activation = tf.nn.relu, kernel_initializer= initializer, bias_initializer=initb))
model3.add(tf.keras.layers.Dense(300, activation= tf.nn.relu,kernel_initializer=initializer,bias_initializer=initb))

model3.add(tf.keras.layers.Dense(6, activation = tf.nn.softmax))

model3.compile(optimizer='SGD',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

model3.fit(x_f_train, y_f_train, epochs=100, batch_size = 10)
score,acc = model3.evaluate(x_f_test, y_f_test)
```
